```python
import sqlite3

connection=sqlite3.connect('classroomDB.db')
cursor=connection.cursor()

create_table = """
CREATE TABLE classroom (
    student_id interger primery key,
    name varchar(30),
    gender char(1),
    physics_marks integer,
    chemistry_marks integer,
    mathematics_marks integer
   
);
"""
cursor.execute(create_table)
connection.commit()
connection.close()
```

```python
classroom_data=[(1,"Raj","M",50,60,70),
                (2,"man","M",80,60,70),
                (3,"femi","F",90,70,80)   ]
connection=sqlite3.connect('classroomDB.db')
cursor = connection.cursor()
for row in classroom_data:
    insert_query="""
                    insert into classroom (student_id,name,gender,physics_marks,chemistry_marks,mathematics_marks)
                    values 
                    ({0},"{1}","{2}",{3},{4},{5})""".format(row[0],row[1],row[2],row[3],row[4],row[5])
    cursor.execute(insert_query)
connection.commit()
connection.close()
```

```python
import sqlite3
connection=sqlite3.connect("classroomDB.db")
cursor=connection.cursor()
cursor.execute('select * from classroom;')
result=cursor.fetchall()
print(result)
for row in result:
    print(row)
connection.close()
```
